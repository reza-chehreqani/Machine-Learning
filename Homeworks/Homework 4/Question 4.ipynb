$$
H(Y) = -\sum_{y} P(Y=y) \log_{2} P(Y=y), \quad H(Y|X_{i}) = \sum_{x} P(X_{i}=x) H(Y|X_{i}=x)
$$

```python
import numpy as np

def calculate_entropy(data, attribute):
    """Calculates the entropy of an attribute based on the goal column."""

    # Group by the input attribute and get the distribution of the goal column
    distributions = data.groupby(attribute)['WillWait'].value_counts(normalize=True).unstack()

    probabilities = data[attribute].value_counts(normalize=True)

    entropy = np.sum(probabilities * -np.sum(distributions * np.log2(distributions), axis=1))
    return entropy

# Calculate entropy for each input attribute
pd.Series({column: calculate_entropy(df, column) for column in df.columns[:-1]})
```

```python
def build_tree(data):
    """Builds a decision tree recursively."""
    labels = data['WillWait'].unique()
    if len(labels) == 1:
        return {'value': labels[0]}

    best_attribute = min(data.columns[:-1], key=lambda x: calculate_entropy(data, x))
    tree = {'attribute': best_attribute, 'children': {}}
    for value in data[best_attribute].unique():
        subset = data[data[best_attribute] == value]
        subtree = build_tree(subset)
        tree['children'][value] = subtree

    return tree

build_tree(df)
```
